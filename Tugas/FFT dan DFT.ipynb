```python
t = np.linspace(0, 10, 1000)
y = np.sin(30*t)
y_fft = np.fft.fft(y)
fig, ax = plt.subplots()
ax.plot(t,y, label='sin(t)')
ax.plot(t, y_fft, label='FFT of sin(t)')
ax.legend()
plt.show()
```
